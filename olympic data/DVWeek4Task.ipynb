```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/drive/MyDrive/olympic_results.csv")
df.head()
```

```python
sport=[]
medal=[]
rank_equal=[]
country=[]
country_code=[]
position=[]
three_letter_code=[]
slug_game=[]

for i in range(15):
    j=i*1000
    sport.append(df["discipline_title"].iloc[j])
    medal.append(df["medal_type"].iloc[j])
    rank_equal.append(df["rank_equal"].iloc[j])
    country.append(df["country_name"].iloc[j])
    country_code.append(df["country_code"].iloc[j])
    position.append(df["rank_position"].iloc[j])
    three_letter_code.append(df["country_3_letter_code"].iloc[j])
    slug_game.append(df["slug_game"].iloc[j])
    
    
print(sport)
print(medal)
print(rank_equal)
print(country)
print(country_code)
print(position)
print(three_letter_code)
print(slug_game)
```

```python
ordinal=[]
for i in range(15):
    x=[]
    x.append(medal[i])
    
    ordinal.append(x)
print(ordinal)
```

```python
binary=[]
for i in range(15):
  x=[]
  x.append(rank_equal[i])
  binary.append(x)
print(binary)
```

```python
quantitative=[]
for i in range(15):
    x=[]
    x.append(position[i])
    
    quantitative.append(x)
print(quantitative)
```

```python
nominal=[]
for i in range(15):
    x=[]
    x.append(sport[i])
    x.append(slug_game[i])
    x.append(three_letter_code[i])
    x.append(country[i])
    x.append(country_code[i])
    nominal.append(x)
print(nominal)
```

```python
data=[]
for i in range(15):
  x=[]
  x.append(sport[i])
  x.append(slug_game[i])
  x.append(three_letter_code[i])
  x.append(country[i])
  x.append(country_code[i])
  x.append(medal[i])
  x.append(rank_equal[i])
  x.append(position[i])
  data.append(x)
print(data)
```

```python
cell_colors=[]
for i in range(15):
  temp=[]
  for j in range(8):
    temp.append('')
  cell_colors.append(temp)
for i in range(15):
  for j in range(8):
    if j==0:
      cell_colors[i][j]="lightblue"
    elif j==1:
      cell_colors[i][j]="lightblue"
    elif j==2:
      cell_colors[i][j]="lightblue"
    elif j==3:
      cell_colors[i][j]="lightblue"
    elif j==4:
      cell_colors[i][j]="lightblue"
    elif j==5:
      cell_colors[i][j]="lightgreen"
    elif j==6:
      cell_colors[i][j]="red"
    else:
      cell_colors[i][j]="yellow"
```

```python
columns=['SPORTS','SLUG_GAME','3_LETTER_CODE','COUNTRY','COUNTRY_CODE','MEDAL_TYPE','RANK_EQUAL','POSITION']
fig, ax = plt.subplots()
ax.set_axis_off()
t=plt.table(cellText=data,colLabels=columns,colColours=["lightblue","lightblue","lightblue","lightblue","lightblue","lightgreen","red","yellow"],cellLoc="center",rowLoc="center",loc="center",cellColours=cell_colors)
    
t.scale(5, 5)
t.set_fontsize(50)
plt.show()
```
